3.1. Створіть зводну таблицю, яка покаже, скільки покупок (інвойсів) було зроблено в кожній з філій (`Branch`) за різними категоріями товарів. Запишіть таблицю в змінну `invoices_by_category` і виведіть змінну на екран.

```python
invoices_by_category = df.pivot_table(index='Branch', columns='Product line', values='Invoice ID', aggfunc='count')
```

Маючи дані про середній чек покупки, можна проанаізувати хто з жінок / чоловіків має більшу купівельну спроможність в різних філіалах. Це може вказати на те, які товари будуть більш пріоритетні в даному магазині. (для чоловіків це техніка, для жінок - товари для здоров', краси, побуту - і в залежності від цих данних акцентувати асортимент).

Питання №2: Визначити який середній чек покупки для кожного філіалу в залежності від гендеру.

```python
mean_unit_price = df.pivot_table(index='Branch', columns='Gender', values='Unit price', aggfunc='mean')
mean_unit_price
```

```python
df.groupby(by=['Branch', 'Gender'])['Unit price'].mean()
```

На основі цих даних можна помітити те що в філіалі А середній чек покупки в жінок домінує хоч і кількість покупок здійснило більше чоловіків. Отже жінки купували більш дороговартісні товари. Можна розширити лінійку жіночого одягу і аксесуарів більш преміальними брендами.
А в філіалі С, навпаки, чоловіки витратили більше, хоча більша кількість покупок була здійснена жінками. Тому в цьому випадку треба так само розшири асортимент для чоловіків більш перміальними позиціями.
